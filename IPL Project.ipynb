```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv("IPLData2.csv")
```

```python
data
```

```python
batter= data[data["Player_Type"]=="Batter"]
new_batter=batter[batter["Capped"]==1]
capped_batters= new_batter[["Player Name",
                                "Team",
                                "Average",
                                "Matches_Played",
                                "Runs",
                               "Strike_Rate"]]
```

```python
allrounder = data[data["Player_Type"]== 'Allrounder']
new_allrounders = allrounder[allrounder["Capped"]==1]
capped_rounders = new_allrounders[["Player Name",
                                  "Team",
                                  "Matches_Played",
                                  "Runs",
                                   "Average",
                                  "Strike_Rate",
                                  "Bowling_average",
                                  "Catches",
                                   "Economy",
                                  'Stumps',
                                  'Wickets','Run_outs']]
```

```python
keeper=data[data["Player_Type"] =="Keeper"]
new_keeper = keeper[keeper["Capped"] == 1]
capped_keepers= new_keeper[["Player Name",
                            "Team",
                            "Average",
                            "Matches_Played",
                            "Run_outs",
                            "Stumps",
                           "Catches"]]
```

```python
#Filling null or nan value with 0
capped_batters = capped_batters.fillna(0)
capped_rounders = capped_rounders.fillna(0)
capped_keepers = capped_keepers.fillna(0)
```

```python
#Analzysing Batters Data

top_batters=capped_batters[capped_batters["Average"]>=32.0]

top_batters_Average = top_batters.sort_values("Average", ascending = False)
top_batters_Strike_Rate = top_batters.sort_values("Strike_Rate", ascending = False)
top_batters_Runs = top_batters.sort_values("Runs", ascending = False)
top_batters_Matches_Played = top_batters.sort_values("Matches_Played", ascending = False)
```

```python
#Analzysing All_rounders Data

top_rounders =capped_rounders[capped_rounders["Strike_Rate"]>=140]

top_rounders_Average=top_rounders.sort_values("Average",ascending = False)
top_rounders_Runs=top_rounders.sort_values("Runs",ascending = False)
top_rounders_Matches_Played=top_rounders.sort_values("Matches_Played",ascending = False)
top_rounders_Bowling_average=top_rounders.sort_values("Bowling_average",ascending = False)
top_rounders_Wickets=top_rounders.sort_values("Wickets",ascending = False)
top_rounders_Economy=top_rounders.sort_values("Economy",ascending = False)
top_rounders_Catches=top_rounders.sort_values("Catches",ascending = False)
```

```python
#Analzysing Keepers Data

top_keepers =capped_keepers[capped_keepers["Average"]>=25]


top_keepers_Stumps =top_keepers.sort_values("Stumps", ascending = False)
top_keepers_Stumps =top_keepers.sort_values("Catches", ascending = False)
top_keepers_Matches_Played =top_keepers.sort_values("Matches_Played", ascending = False)
top_keepers_Matches_outs =top_keepers.sort_values("Run_outs", ascending = False)
top_keepers_Avg =top_keepers.sort_values("Average", ascending = False)
```
